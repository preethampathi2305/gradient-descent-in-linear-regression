```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
x,y=make_regression(n_samples=400,n_features=1,n_informative=1,noise=1.8,random_state=11)
x.shape,y.shape
```

```python
y=y.reshape((-1,1))
y.shape
```

```python
x[:5],y[:5]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
x=(x-x.mean())/x.std()
x[:5]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
ones=np.ones((x.shape[0],1))
newx=np.hstack((x,ones))
newx.shape
```

```python
def hypothesis(x,theta):
    return np.dot(x,theta)
def calculateTheta(x,y):
    y=np.mat(y)
    p1=np.dot(x.T,x)
    p2=np.dot(x.T,y)
    theta=np.linalg.pinv(p1)*p2
    return theta
```

```python
theta=calculateTheta(newx,y)
theta
```

```python
plt.scatter(x,y)
plt.plot(x,hypothesis(newx,theta),color='red')
plt.show()
```

```python
len(x)
```

```python
def error(x,y,theta):
    m=x.shape[0]
    ycap=hypothesis(x,theta)
    e=0
    for i in range(len(y)):
        e+=(y[i]-ycap[i])**2
    return e/m
error(newx,y,theta)
```
